```python
import akshare as ak 

import pandas as pd
import numpy as np

import os 
from datetime import datetime
from tqdm import tqdm
```

```python
# load the hongli list
hs = pd.read_csv('../data/input/hongli_list_20251213.csv')
hs['code'] = hs['code'].astype(str).str.zfill(6)

STOCK_CODES = hs['code'].tolist()
print(len(STOCK_CODES))
```

```python
today = pd.to_datetime("today").strftime("%Y%m%d")

os.makedirs(f"../data/input/price-data/{today}", exist_ok=True)
```

```python
# check financial data already loaded
STOCK_LOADED = os.listdir(f'../data/input/price-data/{today}/')
STOCK_LOADED = [stock[13:19] for stock in STOCK_LOADED]
 
len(STOCK_LOADED)
```

```python
MISSED_CODE = [stock for stock in STOCK_CODES if stock not in STOCK_LOADED]
len(MISSED_CODE)
```

```python
STOCK_SYMBOLS_1 = ["sh" + code if code.startswith("6") 
                   else "bj" + code if code.startswith("9") 
                   else "sz" + code 
                   for code in MISSED_CODE]
len(STOCK_SYMBOLS_1)
```

```python
today = pd.to_datetime("today").strftime("%Y%m%d")

while len(STOCK_SYMBOLS_1) > 0:
    print(len(STOCK_SYMBOLS_1))
    for symbol in tqdm(STOCK_SYMBOLS_1):
        try:
            price_df = ak.stock_zh_a_daily(symbol=f"{symbol}", start_date="20101231", end_date="20251231", adjust="")
            price_df = price_df[['date', 'open', 'high', 'low', 'close']]
            price_df.columns = ['report_date', 'open', 'high', 'low', 'close']
            price_df.to_csv(f"../data/input/price-data/{today}/price_data_{symbol}_{today}.csv", index=False)
            STOCK_SYMBOLS_1.remove(symbol)
        except: 
            continue
```
