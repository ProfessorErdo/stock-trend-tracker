```python
import akshare as ak
```

```python
codes = ['SH600519', 'SZ000858', 'SH600938', 'SZ000333', 'SH601088', 'SZ300866', 'SH600900', 'SH600036', 
         'SH601318', 'SZ300750', 'SH603993', 'SH601899', ]
```

```python
for code in codes:
    
    # extract the pe_ttm and stock_name
    stock_individual_spot_xq_df = ak.stock_individual_spot_xq(symbol=code)
    pe_ttm = stock_individual_spot_xq_df.iloc[34, 1]
    stock_name = stock_individual_spot_xq_df.iloc[13, 1]
    
    # extract the roe in average of 3 years
    stock_zh_dupont_comparison_em_df = ak.stock_zh_dupont_comparison_em(symbol=code)
    stock_zh_dupont_comparison_em_df.index = stock_zh_dupont_comparison_em_df['代码']
    roe_3a = stock_zh_dupont_comparison_em_df.loc[code[-6:], "ROE-3年平均"]
    
    # extract the roe_ttm
    stock_financial_abstract_ths_df = ak.stock_financial_abstract_ths(symbol=code[-6:], indicator="按报告期")
    stock_financial_abstract_ths_df.index = stock_financial_abstract_ths_df['报告期']
    periods = 3 # the current financial report period
    roe_1 = float(stock_financial_abstract_ths_df[['净资产收益率']].iloc[-periods-1].values[0].strip('%'))
    roe_2 = float(stock_financial_abstract_ths_df[['净资产收益率']].iloc[-periods-1-(4-periods)].values[0].strip('%'))
    roe_gap = roe_1 - roe_2
    roe_report = float(stock_financial_abstract_ths_df[['净资产收益率']].iloc[-1].values[0].strip('%'))
    roe_ttm = roe_report + roe_gap
    
    # calculate the pr_3a and pr_ttm
    pr_3a = pe_ttm / roe_3a
    pr_ttm = pe_ttm / roe_ttm
    
    print(f"{stock_name:<6} | pe-ttm: {pe_ttm:<10} | roe_3a: {roe_3a:<10} | pr_3a: {pr_3a:<10.2f} | \
        roe_ttm: {roe_ttm:<10.2f} | pr_ttm: {pr_ttm:<10.2f}")
```
